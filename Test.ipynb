```python
def fib(nterms):
    a, b = 0, 1
    for i in range(nterms):
        print(a)
        a, b = b, a+b
    
fib(10)
```

```python
def rec_fib(n):
    if n <= 1:
        return n
    else:
        return rec_fib(n-1)+rec_fib(n-2)

for i in range(10):
    print(rec_fib(i))
```
